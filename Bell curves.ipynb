```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data (replace this with your dataset)
data = [20, 25, 35, 40, 30, 20]

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Set up the x values for the bell curve
x = np.linspace(min(data), max(data), 100)

# Calculate the probability density function (PDF) for the normal distribution
pdf = norm.pdf(x, mean, std_dev)

# Plot the bell curve
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', label='Bell curve (Normal distribution)')
plt.fill_between(x, pdf, where=(x < mean - 1.96 * std_dev) | (x > mean + 1.96 * std_dev), color='skyblue', alpha=0.5)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Bell Curve with Two-Tailed Critical Regions')
plt.grid(True)
plt.legend()

# Calculate critical values for two-tailed test (alpha = 0.05)
lower_critical = mean - 1.96 * std_dev
upper_critical = mean + 1.96 * std_dev

# Shade the two critical regions
plt.fill_between(x, pdf, where=(x < lower_critical), color='orange', alpha=0.5, label='Critical Region')
plt.fill_between(x, pdf, where=(x > upper_critical), color='orange', alpha=0.5)

# Show legend with only the relevant labels
handles, labels = plt.gca().get_legend_handles_labels()
relevant_handles = [handles[0], handles[-1]]
relevant_labels = [labels[0], labels[-1]]
plt.legend(relevant_handles, relevant_labels)

plt.show()
```
